```python
import matplotlib.pyplot as plt
from statistics import*
import pandas as pd
import math
from google.colab import drive
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
grupo1=pd.read_csv('grupo1.csv')
grupo2=pd.read_csv('grupo2.csv')
grupo3=pd.read_csv('grupo3.csv')
grupo1.describe()
grupo2.describe()
grupo3.describe()
```

```python
plt.hist(grupo1['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones grupo1')
```

```python
plt.hist(grupo2['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones grupo2')

```

```python
plt.hist(grupo3['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones grupo3')
```

```python
x = grupo1[['HorasEstudio']]
y = grupo1['Calificacion']
model = LinearRegression()
model.fit(x, y)
pendiente = model.coef_[0]
intercepto = model.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
plt.scatter(grupo1['HorasEstudio'], grupo1['Calificacion'], color='blue', label='Datos')
plt.plot(grupo1[['HorasEstudio']], model.predict(x), color='red', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones Grupo1')
plt.legend()
plt.show()

coef_corr, _ = pearsonr(grupo1['HorasEstudio'], grupo1['Calificacion'])
print("Coeficiente de correlación de Pearson:",coef_corr)
```

    Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales:

El coeficiente de regresión (pendiente) indica cuánto cambia la calificación final por cada hora adicional de estudio. Por ejemplo, para el grupo 1, el coeficiente de 5.01 sugiere que por cada hora adicional de estudio, la calificación final aumenta en promedio 5.01 puntos.


    ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?

El coeficiente de correlación de Pearson varía entre -1 y 1. Un valor positivo indica una relación directa, mientras que un valor negativo indica una relación inversa. La magnitud del coeficiente indica la fuerza de la relación:
0.7 a 1 (o -0.7 a -1) indica una relación fuerte.
0.3 a 0.7 (o -0.3 a -0.7) indica una relación moderada.
0 a 0.3 (o 0 a -0.3) indica una relación débil.
Para el grupo 1, un coeficiente de 0.85 sugiere una relación fuerte y positiva entre las horas de estudio y las calificaciones.


    ¿El modelo de regresión lineal es estadísticamente significativo?

La significancia estadística del modelo de regresión lineal se puede evaluar formalmente mediante pruebas adicionales (como la prueba t para la pendiente y el valor p). Sin embargo, un alto coeficiente de correlación sugiere que la relación entre las variables es probablemente significativa. En el contexto dado, altos valores de los coeficientes de correlación indican que el modelo es significativo.


    ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?

La confiabilidad del modelo se evalúa por la fuerza de la correlación y otros factores como el error estándar de los estimadores. Un alto coeficiente de correlación (cercano a 1) indica un modelo confiable. Por ejemplo, en el grupo 1, con un coeficiente de 0.85, el modelo es bastante confiable para predecir las calificaciones en función de las horas de estudio. Sin embargo, siempre es importante considerar que la correlación no implica causalidad y que otros factores no incluidos en el modelo también pueden afectar las calificaciones.
